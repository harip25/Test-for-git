```python
import pandas as pd
import nltk 
from nltk.corpus import stopwords 
from nltk import tokenize,RegexpTokenizer
df = pd.read_csv(r'F:\New folder\Due Date (2010)\New folder (5)\Downloads\IMDB Dataset.csv')
dataset = df.values
print(df)
```

```python
Rlist=list()
lines=df['review'].values.tolist()
for line in lines:
    tok=RegexpTokenizer(r'\w+')
    tokens=tok.tokenize(line)
    words=[i.lower() for i in tokens]
    Stopwords=set(stopwords.words("english"))
    words=[i for i in words if not i in Stopwords] 
    Rlist.append(words)
```

```python
import gensim
model=gensim.models.Word2Vec(sentences=Rlist[0:30000],size=100,workers=4)
words1=list(model.wv.vocab)
len(words1)
```
